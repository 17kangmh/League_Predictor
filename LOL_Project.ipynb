```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('high_diamond_ranked_10min.csv')
data
```

```python
data.dropna()
```

```python
data = data.drop(['redTotalGold', 'redTotalExperience', 'redTotalMinionsKilled', 'redTotalJungleMinionsKilled'], axis = 1)
data = data.drop(['blueTotalGold', 'blueTotalExperience', 'blueTotalMinionsKilled', 'blueTotalJungleMinionsKilled'], axis = 1)
data = data.drop(['gameId'], axis = 1)
data
```

```python
data.hist(figsize = (15,8))
```

```python
plt.figure(figsize=(25, 10))
heatmap = sns.heatmap(data.corr(), annot=True)
heatmap
```

```python
data = data.drop(['blueAvgLevel', 'redAvgLevel'], axis = 1)
```

```python
data = data.drop(['redFirstBlood', 'redKills', 'redDeaths'], axis = 1)
```

```python
data = data.drop(['redGoldDiff', 'redExperienceDiff', 'blueExperienceDiff', 'blueGoldDiff'], axis = 1)
```

```python
data = data.drop(['redGoldPerMin', 'redCSPerMin', 'blueGoldPerMin', 'blueCSPerMin'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['blueWins'], axis = 1)
y = data['blueWins']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

logReg = LogisticRegression(max_iter=1000)
logReg.fit(X_train, y_train)
logPrediction = logReg.predict(X_test)
logRegScore = accuracy_score(y_test, logPrediction)
logRegScore
```

```python
coefficients = logReg.coef_[0]
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': np.abs(coefficients)})
feature_importance = feature_importance.sort_values('Importance', ascending=True)
feature_importance.plot(x='Feature', y='Importance', kind='barh', figsize=(10, 6))
```

```python
dataClone = data[['blueWins', 'blueKills', 'blueDeaths', 'blueTowersDestroyed', 'blueDragons']]
```

```python
dataClone
```

```python
dataClone = dataClone.rename(columns={'blueDeaths': 'redKills'})
```

```python
dataClone
```

```python
from sklearn.model_selection import train_test_split
X = dataClone.drop(['blueWins'], axis = 1)
y = dataClone['blueWins']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_Scaler_train = scaler.fit_transform(X_train)
X_Scaler_test = scaler.transform(X_test)

logReg = LogisticRegression(max_iter=1000)
logReg.fit(X_Scaler_train, y_train)
logPrediction = logReg.predict(X_Scaler_test)
```

```python
from sklearn.metrics import mean_squared_error

error = np.sqrt(mean_squared_error(y_test, logPrediction))
error
```

```python
from sklearn.tree import DecisionTreeRegressor

decTreeReg = DecisionTreeRegressor(random_state=0)
decTreeReg.fit(X_Scaler_train, y_train)
decTreeRegPred = decTreeReg.predict(X_Scaler_test)

```

```python
error = np.sqrt(mean_squared_error(y_test, decTreeRegPred))
error
```

```python
from sklearn.ensemble import RandomForestRegressor

randForReg = RandomForestRegressor(random_state=0)
randForReg.fit(X_Scaler_train, y_train)
randForRegPred = randForReg.predict(X_Scaler_test)

error = np.sqrt(mean_squared_error(y_test, randForRegPred))
error
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2, 4, 6, 8, 10]
parameters = {'max_depth': max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring = 'neg_mean_squared_error')
gs.fit(X_train,y_train)

newRegressor = gs.best_estimator_

newRegressor.fit(X_Scaler_train, y_train)
newRegPred = newRegressor.predict(X_Scaler_test)

error = np.sqrt(mean_squared_error(y_test, newRegPred))
error
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2, 4, 6, 8, 10]
parameters = {'max_depth': max_depth}

regressor = RandomForestRegressor(random_state=0)
gsFor = GridSearchCV(regressor, parameters, scoring = 'neg_mean_squared_error')
gsFor.fit(X_train,y_train)

newForRegressor = gsFor.best_estimator_

newForRegressor.fit(X_Scaler_train, y_train)
newForRegPred = newForRegressor.predict(X_Scaler_test)

error = np.sqrt(mean_squared_error(y_test, newForRegPred))
error
```
